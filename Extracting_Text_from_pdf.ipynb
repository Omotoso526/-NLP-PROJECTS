```python
import PyPDF2 as pdf
```

```python
file = open('/content/Master-of-Science-in-Business-MSBS-2024-1.pdf', 'rb')
```

```python
pdf_reader = pdf.PdfReader(file)
```

```python
pdf_reader
```

```python
help(pdf_reader)
```

```python

num_pages = len(pdf_reader.pages)


print(f"The PDF has {num_pages} pages.")
```

```python
page1 = pdf_reader.pages[7]
page1.extract_text()
```

```python
pdf_writer = pdf.PdfWriter()
pdf_writer.add_page(page1)
```

```python
output = open('pages_pdf', 'wb')
pdf_writer.write(output)
output.close()
```
